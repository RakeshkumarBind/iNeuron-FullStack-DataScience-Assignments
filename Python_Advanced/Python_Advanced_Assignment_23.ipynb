# Q3.
What is a histogram, exactly? Name a numpy method for creating such a graph.

<b>Answer:</b><br>
A histogram is a graphical representation of the distribution of a dataset. It provides a visual summary of the frequency or occurrence of values within a specific range or "bin" in the dataset. In a histogram, the x-axis represents the range of values or bins, and the y-axis represents the frequency or count of data points falling into each bin.<br>

In NumPy, you can create a histogram using the `numpy.histogram()`

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
data = np.random.randn(1000)  # Create an array of 1000 random data points

# Create a histogram
hist, bins = np.histogram(data, bins=10)  # 'bins' specifies the number of bins

# Plot the histogram
plt.hist(data, bins=10, alpha=0.5, color='b', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# Q4.
If necessary, how do you change the aspect ratios between the X and Y axes?

<b>Answer:</b><br>
Using `plt.figure(figsize=(width, height))`

```python
import matplotlib.pyplot as plt

# Create a sample plot
x = [1, 2, 3, 4]
y = [2, 4, 1, 3]

# Create a figure with a specific width and height (adjust to control aspect ratio)
plt.figure(figsize=(60, 40))  # Example: Width is 6 units, height is 3 units

plt.plot(x, y)
plt.title("Custom Aspect Ratio")

# Show the plot
plt.show()
```

# Q5.
Compare and contrast the three types of array multiplication between two numpy arrays: dot
product, outer product, and regular multiplication of two numpy arrays.

<b>Answer:</b><br>
Dot Product: Used for matrix multiplication and linear algebraic operations. Requires compatible dimensions, and the result can be a scalar or multi-dimensional array.<br>

Outer Product: Generates a matrix where each element is the product of an element from the first array and an element from the second array. Does not require compatible dimensions between the arrays.<br>

Regular (Element-wise) Multiplication: Multiplies corresponding elements in two arrays element by element. Requires compatible dimensions, and the result is an array of the same shape as the input arrays.<br>

# Q7.
Can string data be stored in numpy arrays? If so, list at least one restriction that applies to this
data.

<b>Answer:</b><br>Yes, an array can store the string. The limitation which imposed on the string data is, whenever we store the data of string dtype then we should keep in mind that the string which is having the maximum length will become the limit.

```python
# importing numpy as np
import numpy as np
  
# Create the numpy array
name = np.array(['Rakesh','Harit','Vineet','Lalsha',''])
  
# Print the array
print(name)

# Assign 'SunitaBind' at the place of missing value
name[name == ''] = 'SunitaBind'
  
# Print the modified array
print(name)
```
