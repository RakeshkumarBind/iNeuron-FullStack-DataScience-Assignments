# Q1.
What is a probability distribution, exactly? If the values are meant to be random, how can you
predict them at all?

<b>Answer:</b><br>A probability distribution is a mathematical function that describes the likelihood of obtaining certain values for a random variable. Probability distributions are used to model the uncertainty associated with random phenomena, such as rolling dice, flipping coins, or measuring the heights of people.<br>

A probability distribution does not predict the exact value of a random variable, but rather specifies the relative likelihood of different values occurring. For example, if we toss a fair six-sided die, the probability distribution for the outcome would specify that each of the six possible outcomes (1, 2, 3, 4, 5, or 6) has an equal chance of occurring, i.e., a probability of 1/6. This means that we cannot predict with certainty what number will appear on the die, but we can say that some numbers are more likely to appear than others.<br>

There are many ways to define a probability distribution, depending on the nature of the random variable being modeled. Some common methods include:<br>

1.`Discrete distributions`: These describe the probabilities of discrete outcomes, such as the roll of a die or the number of heads obtained when flipping a coin. Examples include the binomial distribution and the Poisson distribution.<br>
2.`Continuous distributions`: These describe the probabilities of continuous outcomes, such as the height of a person or the time between arrivals in a queuing system. Examples include the normal distribution and the exponential distribution.<br>
3.`Joint distributions`: These describe the probabilities of multiple random variables occurring together. Examples include the joint distribution of two dice rolls or the joint distribution of the height and weight of a population.<br>
4.`Conditional distributions`: These describe the probabilities of a random variable given the occurrence of another random variable. Examples include the conditional distribution of the second dice roll given the first roll or the conditional distribution of the height of a person given their gender.<br>

Latter is considered good enough as they have sort of uniform distribution. # Q3.
What are the two main factors that influence the behaviour of a &quot;normal&quot; probability distribution?

<b>Answer:</b><br>The two main factors that influence the behavior of a "normal" probability distribution are:<br>

1.`Mean`: The mean of a probability distribution is the average value of the random variable being modeled. It is the center point around which the distribution is symmetric. A higher mean shifts the distribution to the right, while a lower mean shifts it to the left.<br>
2.`Standard Deviation`: The standard deviation of a probability distribution measures the spread or dispersion of the data from the mean. A larger standard deviation indicates that the data points are more spread out, resulting in a wider distribution. Conversely, a smaller standard deviation indicates that the data points are closer to the mean, resulting in a narrower distribution.<br>
<br>
Together, the mean and standard deviation fully determine the shape and scale of a normal probability distribution. By changing the mean and/or standard deviation, you can manipulate the position, shape, and spread of the distribution.<br>

# Q6.
What kind of object can be shuffled by using random.shuffle?

<b>Answer:</b><br>
1.`Lists`: random.shuffle(my_list) will shuffle the elements of the list in place.<br>
2.`Tuples`: random.shuffle(my_tuple) will shuffle the elements of the tuple in place.<br>
3.`Arrays`: random.shuffle(my_array) will shuffle the elements of the array in place.<br>
4.`Iterators`: random.shuffle(iter(my_iterator)) will shuffle the elements returned by the iterator.<br>
5.`Strings`: random.shuffle(my_string) will shuffle the characters of the string in place.<br>
6.`Dictionaries`: random.shuffle(my_dict) will shuffle the key-value pairs of the dictionary in place.<br>
7.`Sets`: random.shuffle(my_set) will shuffle the elements of the set in place.<br>
